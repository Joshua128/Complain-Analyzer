**TRAINING A BASELINE MODEL**

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns



train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
validation_data = pd.read_csv("validation.csv")

```

**TF-IDF Vectorizer will be used to train a baseline model**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

train_texts = train_data["text"].tolist()

vectorizer = TfidfVectorizer()

train_tfidf_matrix = vectorizer.fit_transform(train_texts)

#currently have  tf-idf matrix of shape (30000, len(vocab)) 



print(train_tfidf_matrix.shape)

```

**Represents what the general classes were transformed too**

```python
print(vectorizer.get_feature_names_out()[:10])
```

```python
#each  sample now has associated features with tf-idf weights for each word
#Logistic regression can take these inputs and in turn map them to the corresponding labels (clean_product) for classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report


X_train = train_tfidf_matrix
y_train = train_data["labels"]
model = LogisticRegression(max_iter=1000, class_weight="balanced")
model.fit(X_train, y_train)

val_preds = model.predict(vectorizer.transform(validation_data["text"]))
y_val = validation_data["labels"]
label_names = sorted(train_data["labels"].unique())



print(classification_report(
    y_val,
    val_preds,
    target_names=label_names
))

```
